## Determinamos si hay datos nulos
Del siguiente proceso tenemos que hay valores nulos en las columnas Saving accounts y Checking account lo cual se puede deber a que hay personas que solicitan prestamos al banco pero no necesariamente tienen cuentas con el banco.

Vamos a sustituir los valores nulos de las columnas 'Saving accounts' y 'Checking account' por 'without account'.

```python
type_account = ['Saving accounts', 'Checking account']
for i in type_account:
    df[i].fillna('without account', inplace= True)
```
